### 문제 1
##### 다음 6 개의 Expression에 대해 Evaluation 결과 값을 출력하고, 해당 결과가 나온 이유에 대해 설명하시오

```python
# 1번

print(1 and 2 and 3 and 4)
```

#### 1. 1 and 2 and 3 and 4
##### 기본적으로 왼쪽에서 오른쪽으로 조건을 확인하는데, and의 경우는 모두 True일 때, True이기 때문에, False가 있는지, 확인해야한다. 따라서, False 값이 보일때까지, and 조건을 지나서 마지막 숫자까지 확인을 해서 4가 나오게 되는 것이다.


```python
# 2번

print(1 or 2 or 3 or 4)
```

#### 2. 1 or 2 or 3 or 4
##### or 같은 경우, 하나라도 True라면 True인 규칙을 가지고 있다. 따라서 or 기준으로 왼쪽의 값만이 참이라면, 뒤는 볼 것 없이 참이라는 것을 의미한다. 2번 같은 경우 1이 True값에 해당하므로, 뒤에 있는 숫자들까지 확인할 필요 없이 1이 나오게 되는 것이다.

```python
# 3번

print(1 and 2 or 3 and 4)
```

#### 3. 1 and 2 or 3 and 4
##### 1번과 2번의 설명을 그대로 따르면 이유를 쉽게 알 수 있다. 먼저 가장 왼쪽의 and를 통해 1과 2를 모두 확인한 후, or을 만나게 되는데, 1 and 2는 2란 값으로 참을 뜻한다. 따라서 or 오른쪽의 있는 수들은 확인하지 않고 2가 나오게 된다.

```python
# 4번

print((1 and 2) or (3 and 4))
```

#### 4. (1 and 2) or (3 and 4)
##### 3번에서 and 부분에 우선순위를 가르는 괄호가 추가되었다. 이를 연산을 하면, 1 and 2에서 2가, 3 and 4에서는 4가 나오므로 결국 2 or 4와 같은 동작을 하는 것이다. 따라서 2가 나온다.

```python
# 5번

print(1 or 2 and 3 or 4)
```

#### 5. 1 or 2 and 3 or 4
##### 역시 맨 왼쪽부터 or을 확인하게 되는데, 역시 오른쪽의 값과 전혀 상관없이 왼쪽이 1로 True임으로 1이 나오게 된다.

```python
# 6번

print((1 or 2) and (3 or 4))
```

#### 6. (1 or 2) and (3 or 4)
##### 4번과 마찬가지로 괄호의 연산들을 먼저 취하고 생각하면, 1 and 3과 동일하게 동작하는 걸 알 수 있다. 이 경우에는 3이 나오게 되는 것이다.

### 문제 2

##### 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.
###### -인자로 전달하는 문자열은 경로만 들어간다고 가정한다.
###### -각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
###### - 반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.


```python
def div_path(s):
    index = s.rfind('/')
    l = [s[:index], s[index+1:]]
    return l

print(div_path('/usr/local/bin/python'))
print(div_path('/home/chulsoo/test.txt'))
```

##### 함수형태를 만들어 인자 s를 받는다. 이때 s는 문자열이라는 조건이 있으므로, 예외는 생각할 필요가 없다. rfind함수는 문자열에 사용가능한 함수로, 인자의 문자를 문자열 내에서 찾아 가장 마지막으로 찾은 인덱스 값을 숫자로 반환하는 함수이다. 따라서  index = s.rfind('/')에서 index에는 가장 마지막(오른쪽) '/'가 들어있는 인덱스 값이 담기는 것이다. 따라서 이를 기준으로 문자열을 슬라이스하여, 리스트에 넣고 리턴해주었고 예시의 결과값이 동일하게 출력되는 것을 확인할 수 있었다.

### 문제 3
##### 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
###### - 집합 자료형 set은 사용하지 않는다.
###### - 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
###### - 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
###### - 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
###### - 반환하는 리스트는 정렬이 되어 있어야 한다.

```python
def list_union(lista, listb):
    ml = lista + listb
    l = []
    for v in ml:
        if v not in l:
            l.append(v)
    l.sort()
    return l


print(list_union([1, 2, 3], [1, 2, 4]))
print(list_union([-10, -5, 0, -1], [100, 9, 0, 9]))
print(list_union([0, 1, 2], [0, 1, 2]))
```

##### 먼저 인자로 들어온 리스트 두개를 병합한 ml 리스트를 만든다. 그리고 빈 리스트 l을 만들다. 그다음 ml의 값들을 순회하는 반복문을 통하여, ml의 값이, 리스트 l 안에 값중에 없다면 리스트 l에 해당 ml 값을 추가해주는 방식으로 중복되는 숫자를 제거하여 리스트 l을 만들었다. 그후 정렬 함수인 sort 후에 반환하면, 문제를 해결할 수 있다.

### 문제 4
##### 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '교집합'을 반환하는 함수 list_intersection(lista, listb)와 '차집합'을 반환하는 함수 list_difference(lista, listb)를 작성하시오.
###### - 모든 가정과 조건은 3번 문제와 동일하다

```python
def list_intersection(lista, listb): 
    l = []
    for v in listb:
        if v in lista and v not in l:
            l.append(v)
    l.sort()
    return l
  
    
print(list_intersection([1, 2, 3], [1, 2, 4, 3]))
```

##### 교집합) 빈리스트 l을 만든다. 그리고 listb의 값들을 순회하는 반복문에서 listb의 값인 v가 lista에 있으며, v가 새 리스트 l에 존재하지 않을때, v값을 l에 추가해주었다. 첫번째 조건은 교집합을 찾아주는 것이고 두번째 조건은 중복을 제거해주는 역할을 한다. 여기서 나온 리스트 l을 반환하면 위와같이 교집합의 기능을 하는 것을 알 수 있다.

```python
def list_difference(lista, listb):
    l = lista
    for v in listb:
        if v in lista:
            l.remove(v)
    l.sort()
    return l

print(list_difference([1, 2, 3, 4, 5], [1, 2, 3]))
```

##### 차집합) 차집합은 앞에 있는 집합에서 뒤에있는 집합의 원소들을 제거해주는 걸 전제로 코드를 작성하였다. 따라서 lista를 복사한 리스트 l을 만들고 listb의 원소를 순회하는 반복문에서 만약 listb의 원소인 v가 lista에 존재한다면, l에서 v 원소를 제거해주는 것으로 차집합을 구하였다. 역시 잘 동작되는 것을 확인하였다.

### 문제 5
##### 두 개의 양의 정수를 인자로 받아서 해당 범위안에 있는 소수(prime number)의 리스트를 출력하는 함수 print_primenumber(a, b)를 작성하시오.
###### - 이미 구현된 모듈이나 라이브러리는 사용하지 않는다.
###### - 함수의 반환(return)은 없으며 함수 내의 끝에서 .format()을 사용하여 다음과 같이 출력한다.
###### print_primenumber(2, 10)
###### 2 이상 10 이하 정수 중 소수 리스트: [2, 3, 5, 7]

```python
def is_prime(a):
  if a < 2:
    return False
  for i in range(2, a):
    if a % i == 0:
      return False
  return True

def print_primenumber(a, b):
    l = []
    for i in range(a, b):
        if is_prime(i):
            l.append(i)
    print('{0} 이상 {1} 이하 정수 중 소수 리스트:'.format(a,b), l)
    
print_primenumber(2,10)
```

##### 소수를 구하는 알고리즘은 너무나 자명하기 때문에, 설명을 생각하겠다. 출력은 양식에 따라 format 매소드를 사용하여, 위와같이 출력하였다.

### 문제 6
##### 웹 URL로 지정된 웹페이지를 문자열로 가져와 1) 모든 HTML 태그를 제외한 순수 텍스트 문자열만을 걸러내고, 2) 그 순수 텍스트 문자열 안에 존재하는 단어가 총 몇개인지 출력하는 프로그램을 작성하시오.

```python
import requests

def delete_tag(s, start_tag, end_tag):
    start = 0
    end = 0
    
    while True:
        start = s.find(start_tag)
        end = s.find(end_tag) + len(end_tag)
            
        if start == -1:
            break
        
        tmp = s.split(s[start:end], 1)
        s = tmp[0] + '^' + tmp[1]
        
    return s

req = requests.get('http://cse.koreatech.ac.kr')
source = req.text

#print(source)

source = delete_tag(source, "<script", "</script>")
source = delete_tag(source, "<style", "</style>")
source = delete_tag(source, "<", ">")
source = source.replace('\n', "")
source = source.replace('\r', "")

result = []

for v in source.split('^'):
    tmp = v.strip(" ")
    if tmp != "":
        result.append(tmp)

#문자열을 갯수.
print(len(result))

total = 0

for v in result:
    total = len(v.split(" ")) + total

#총 문자의 갯수
print(total)

```

##### delete_tag(s, start_tag, end_tag) 함수는 문자열을 s에 start_tag문자열부터 end_tag문자열까지의 문자를 삭제시켜주는 함수이다. 이를 사용하여, 불필요한 태그들인 script부터 style등의 태그들을 삭제하여 보다 편하게 작업할수 있게 하였다.
##### 그 후, replace, split, strip의 문자열 함수를 사용하여, 불필요한 문자들을 분리 후, 필요한 문자열과, 문자들만 추출하여 그 갯수를 세어주었다.
