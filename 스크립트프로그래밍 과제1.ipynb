```python
# 8번문제

def isPM(n):
    for i in range(2, n):
        if n % i == 0:
            return False 
    return True

num = int(input("정수를 입력하세요: "))

for i in range(1,num):
    if isPM(i) & i != 1:
        print(i)
```

```python
# 10번문제

def appendDelete(s, t, k):
    
    x = min(len(s), len(t))
    
    i = 0
    result = 0
    check = True
    
    for i in range(x):
        if s[i] != t[i]:
            check = False
            break
            
    if check:
        i += 1
        
    result = len(s) + len(t) - 2*i
    
    if k >= result:
        return 'Yes'
    else:
        return 'No'
    


print(appendDelete('hackerhappy', 'hackerrank', 9))
print(appendDelete('aba', 'aba', 0))
print(appendDelete('ashley', 'ash', 2))
```
