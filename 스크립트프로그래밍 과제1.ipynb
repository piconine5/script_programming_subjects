# 6번문제
### while문을 활용하여 1부터 100사이의 짝수의 합을 계산하여 출력하시오.

```python
# 6번문제

count = 1
result = 0

while count <= 100:
    if count % 2 == 0:
        result += count
    count += 1
    
print(result)
```

# 8번문제
### 사용자로 부터 정수를 입력받아서 1부터 그 사이에 존재하는 소수 (Prime number)를 출력하는 파이썬 프로그램을 작성하시오.

```python
# 8번문제

def isPM(n):
    for i in range(2, n):
        if n % i == 0:
            return False 
    return True

num = int(input("정수를 입력하세요: "))

for i in range(1,num):
    if isPM(i) & i != 1:
        print(i)
```

# 10번문제
### 알파벳 소문자만으로 이루어진 string, S, 가 주어지고 아래 2가지 연산을 수행할 수 있다.
#### 추가 : string S의 끝에 알파벳 소문자 하나를 추가한다.
#### 삭제 : string S의 마지막 문자 하나를 삭제한다. empty string에서 삭제 연산을 수행한 결과는 empty string이다.

### 두 개의 string, s와 t, 그리고 정수 k가 주어질 때, 정확히 k번의 연산을 통해 s를 t로 바꿀 수 있는지 판단하는 함수를 작성하시오.

```python
# 10번문제

def appendDelete(s, t, k):
    
    x = min(len(s), len(t))
    
    i = 0
    result = 0
    check = True
    
    for i in range(x):
        if s[i] != t[i]:
            check = False
            break
            
    if check:
        i += 1
        
    result = len(s) + len(t) - 2*i
    
    if k >= result:
        return 'Yes'
    else:
        return 'No'
    


print(appendDelete('hackerhappy', 'hackerrank', 9))
print(appendDelete('aba', 'aba', 0))
print(appendDelete('ashley', 'ash', 2))
```
