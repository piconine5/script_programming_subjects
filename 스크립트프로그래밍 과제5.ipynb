    > \>\>\> c = c + 5  <br/>
    > \>\>\> d = d - 5  <br/>
    > \>\>\> print(c)  <br/>
    > [Count (step: 1)] 17 <br/>

  - ìš”êµ¬ì‚¬í•­ 6. ë‹¤ìŒê³¼ ê°™ì€ ê´€ê³„ì—°ì‚° (>, <, ==)ì´ ìˆ˜í–‰ë  ìˆ˜ ìˆë„ë¡ Counter í´ë˜ìŠ¤ ë‚´ì— ê´€ë ¨ ë©”ì†Œë“œë¥¼ ì¶”ê°€í•˜ì‹œì˜¤.
    > \>\>\> print(c > 10)  <br/>
    > True <br/> 
    > \>\>\> print(d > 10) <br/>
    > False <br/>
    > \>\>\> print(c < 10)  <br/>
    > False <br/> 
    > \>\>\> print(d < 10) <br/>
    > True <br/>
    > \>\>\> print(c == 17)  <br/>
    > True <br/> 
    > \>\>\> print(d != 9) <br/>
    > False <br/>

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    
    def __str__(self):
        return '[Counter (step: {})] {}'.format(self.step, self.count)
    
    def incr(self):
        self.count += self.step
    
    def __call__(self):
        self.incr()

    def __add__(self, other):
        self.count += other
        return self

    def __sub__(self, other):
        self.count -= other
        return self

    def __eq__(self, other):
        return self.count == other

    def __ne__(self, other):
        return self.count != other

    def __gt__(self, other):
        return self.count > other

    def __lt__(self, other):
        return self.count < other
    
    def __ge__(self, other):
        return self.count >= other

    def __le__(self, other):
        return self.count <= other
```

```python
#   ìš”êµ¬ì‚¬í•­ 1
c = Counter(10)
d = Counter(10, 2)

#   ìš”êµ¬ì‚¬í•­ 2
print('print Counter')
print(c)
print(d)
print('\n')

#   ìš”êµ¬ì‚¬í•­ 3
print('Counter increase')
c.incr()
d.incr()

print(c)
print(d)
print('\n')

#   ìš”êµ¬ì‚¬í•­ 4
print('call Counter object')
c()
d()

print(c)
print(d)
print('\n')

#   ìš”êµ¬ì‚¬í•­ 5
print('add and sub')
c = c + 5
d = d - 5

print(c)
print(d)
print('\n')

#   ìš”êµ¬ì‚¬í•­ 6
print('compare Counter')
print('c > 10 : {}'.format(c>10))
print('d > 10 : {}'.format(d>10))
print('c < 10 : {}'.format(c<10))
print('d < 10 : {}'.format(d<10))
print('c == 17 : {}'.format(c==17))
print('d != 9 : {}'.format(d!=9))
```

5. 4ë²ˆ ë¬¸ì œì— ì •ì˜ëœ MySet í´ë˜ìŠ¤ì— ë©”ì†Œë“œë¥¼ ì¶”ê°€í•˜ì—¬ ë‹¤ìŒ ê° ìš”êµ¬ì‚¬í•­ ëª¨ë‘ë¥¼ ë§Œì¡±ì‹œí‚¤ëŠ” ì½”ë”©ì„ ì œì‹œí•˜ì‹œì˜¤.

  - ì •ë‹µì„ ê° ìš”êµ¬ì‚¬í•­ë³„ë¡œ ì…ë ¥í•  í•„ìš” ì—†ì´ ìš”êµ¬ì‚¬í•­ 3ê°œ ì „ì²´ì— ëŒ€í•´ 1ê°œì˜ MySet í´ë˜ìŠ¤ ì •ì˜ ì½”ë“œë¥¼ ì œì‹œí•˜ê³  ì£¼ì„ë“±ìœ¼ë¡œ í•´ë‹¹ ì½”ë“œë¥¼ ì„¤ëª…í•œë‹¤.
  - ìš”êµ¬ì‚¬í•­ 1. | ì—°ì‚°ìœ¼ë¡œ ë‘ ì§‘í•©ì˜ í•©ì§‘í•©ì„ ë°˜í™˜í•œë‹¤.

    > \>\>\> print(u)  <br/>
    > MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} <br/> 
    
  - ìš”êµ¬ì‚¬í•­ 2. & ì—°ì‚°ìœ¼ë¡œ ë‘ ì§‘í•©ì˜ êµì§‘í•©ì„ ë°˜í™˜í•œë‹¤.
    > \>\>\> u = s & t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {2, 3} <br/>

  - ìš”êµ¬ì‚¬í•­ 3. - ì—°ì‚°ìœ¼ë¡œ ë‘ ì§‘í•©ì˜ ì°¨ì§‘í•©ì„ ë°˜í™˜í•œë‹¤.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> u = s - t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {1, 2} <br/>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    def __or__(self, other):
        temp = self + other
        MySet.eliminate_duplicate(temp)
        return MySet(temp)

    def __and__(self, other):
        temp = []
        for i in self:
            for j in other:
                if i == j:
                    temp.append(i)
        MySet.eliminate_duplicate(temp)
        return MySet(temp)

    def __sub__(self, other):
        for i in self:
            for j in other:
                if i == j:
                    self.remove(i)
        return self
        


#   sì™€ tì˜ ìˆ˜ëª…ì´ ifë¬¸ì— ìˆê¸° ë•Œë¬¸ì— ë‹¤ì‹œ ì„ ì–¸
s = MySet([1, 2, 2, 3])
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])


#   í•©ì§‘í•©
u = s | t
print('í•©ì§‘í•© = {}'.format(u))

#   êµì§‘í•©
u = s & t
print('í•©ì§‘í•© = {}'.format(u))

#   í•©ì§‘í•©
s = MySet([1, 2, 3])
t = MySet([3, 4, 5])
u = s - t
print('í•©ì§‘í•© = {}'.format(u))
```
