### 문제5번

##### 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
##### 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
##### 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
import functools

def addall(l) :
    sum = 0
    for i in l:
        if i > 0:
            sum += i
    return sum 

def addall2(l):
    sum = functools.reduce(lambda x,y: x+y, [i for i in l if i > 0])
    return sum


print(addall([1,2,3,4,5,6,7,8,9]))
print(addall2([1,2,3,4,5,6,7,8,9]))
```

##### 1) addall을 for문, 리스트 내포 방식으로 각각 만들었다. 리스트 내포 방식에서는 리스트의 값들을 모두 더해주기 위해, functools의 reduce 메서드를 사용하였고, 간단한 람다 함수를 사용하였다. 편의상 addall2로 네이밍하였다.

```python
def addallodd(l) :
    sum = 0
    for i in l:
        if i % 2:
            sum += i
    return sum 

def addallodd2(l) :
    sum = functools.reduce(lambda x,y: x+y, [i for i in l if i % 2])
    return sum

print(addallodd([1,2,3,4,5,6,7,8,9]))
print(addallodd2([1,2,3,4,5,6,7,8,9]))
```

##### 2) 1과 크게 다르지 않게 만들었고 데이터를 필터링해주는 if문만 조건 수정을 해주었다.

##### 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오
##### length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수
##### Strange라는 숫자를 아래과 같이 정의한다.
##### 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.
##### length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.
##### X는 length(X)로 정확히 나누어 떨어진다.
##### X / length(X) 가 strnage이다.

```python
def Length(n) :
    return len(str(n))
```

##### 먼저 숫자를 받아 그 숫자의 자릿수를 리턴해주는 함수를 만들었다.

```python
def isStrange(num): 
    if Length(num) == 1: return True
    if Length(num) > 1 and num % Length(num) == 0: return isStrange(int(num/Length(num)))
    else: return False
    
print(isStrange(0))
print(isStrange(56))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))
```

##### 문제 조건을 확인했을 때, 먼저 True인 조건을 확인할 수 있었다. 끝까지 숫자가 끝까지 나누어 떨어져서, 자릿수가 1자리가 되면 그 숫자는 strange가 된다.
##### 그럼 나눠주는 조건은, num 자릿수가 1보다 크고, 그 숫자를 자릿수로 나누었을 때, 나누어 떨어진다면, 이는 strange가 될 수도 있는 num임으로 num에서 자릿수를 나눠준 값을 인자로하는 isStrange를 호출하여, 자릿수가 1자리가 나올때까지 반복해주는 재귀함수를 만들어준다. 위 조건이 아니라면 즉, 나누었을때, 나누어 떨어지지 않는다면, 이는 strange가 아니므로 False를 리턴해준다.

### 문제9번

##### 다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.

##### 1번 그룹, 2번 그룹, ... n번 그룹으로 번호가 매겨진 총 n개의 그룹이 존재하고, ith 그룹에는 임의의 팀원 수 ai 명이 있다.
##### 하나의 버스 정류장이 있고, 버스는 오직 한 대만 운영한다. 버스는 버스 정류장에 도착하고 모든 그룹은 버스를 통해 여행하길 원한다.
##### 하지만, 같은 그룹에 속한 팀원들은 분리되기를 원치 않는다. 그래서 그룹의 모든 팀원이 탈 수 있을 때 에만 그 그룹은 버스에 탑승한다.
##### 또한, 이들은 그룹의 번호순으로 여행을 하길 원한다. 예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나, 1번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.
##### 마지막으로, 버스는 빈 좌석이 없을 때 에만 여행을 떠난다.

##### 이 문제에서 각 그룹의 팀원 수가 리스트로 주어질 때, 가능한 모든 버스의 사이즈 x(x명의 사람들이 동시에 탈 수 있음)를 반환하는 함수 solve(a) 를 작성하시오.
##### [주의] 반드시 계산 복잡도를 최대한 낮추세요. 아래 예시 3번을 실행했을 때 결과가 바로 나타날 수 있도록!!

```python
def getDivisor(n):
    n = int(n)
    divisors = []
    divisors_back = [] 

    for i in range(1, int(n**(1/2)) + 1): 
        if (n % i == 0):            
            divisors.append(i)
            if (i != (n // i)): 
                divisors_back.append(n//i)
    return divisors + divisors_back[::-1]

def checkBus(a, x):     # a에는 그룹별 인원수가 들어있는 리스트, x에는 필터한 약수를 인자로 사용
  s = 0;                # 버스가 꽉차있는지 확인해주기 위해 정수형 변수 s를 0으로 초기화
  for i in a:          # 그룹를 기준으로 반복
    if(s + i > x): return False;   # 버스에 탄 사람(s)보다 버스 크기가 작으면 False
    elif(s + i == x): s = 0;        # 버스가 만승이면, 여행을 떠나기 때문에 다시 버스를 비운다.
    else: s += i                    # 버스에 태웠는데, 자리가 남으면, 그룹 수를 더하고, 다음 그룹을 확인
    
  return s == 0   # 리스트를 끝까지 확인했을때, s가 0, 즉 모두 여행을 완료했다면, True를 리턴

def solve(a):
  s = sum(a)
  m = max(a)
  divisors = [i for i in getDivisor(s) if i >= m]
  l = [i for i in divisors if checkBus(a, i)]    

  return l

print(solve([1, 2, 1, 1, 1, 2, 1, 3]))
print(solve([9999] * 10**5))
```

##### getDivisor 함수는 n을 인자로 받아, 약수를 구하고 리스트를 반환하는 함수이다.
##### 우선 문제를 들여다보면, n인승 버스에 꽉찼을 때만, 출발이 가능하며, 그룹 순서는 반드시 지켜야한다. 1차적으로 모든 그룹의 인원수의 약수로 이루어진 인승 버스만 가능하다는 것을 의미한다. 따라서 모든 리스트의 합을 구하고, 이의 약수를 구하였다. 또한 그룹 중 최대값보다 작은 인승의 버스는 애초에 불가능하기 때문에, 약수들 중 최대값보다 작은 약수를 먼저 필터링하였다. 
##### 이렇게 필터링된 약수들을 사용하여, 위 조건을 확인해준다. 코드와 같이 설명하고 싶기 때문에 자세한 설명은 코드 주석으로 하겠다.

```python
import requests


def delete_tag(s, start_tag, end_tag):
    start = 0
    end = 0
    
    while True:
        start = s.find(start_tag)
        end = s.find(end_tag) + len(end_tag)
            
        if start == -1:
            break
        
        tmp = s.split(s[start:end], maxsplit=1)
        s = tmp[0] + '^' + tmp[1]
        
    return s

req = requests.get('http://cse.koreatech.ac.kr')

source = req.text

source = delete_tag(source, "<script", "</script>")
source = delete_tag(source, "<style", "</style>")
source = delete_tag(source, "<", ">")
source = source.replace('\n', "")
source = source.replace('\r', "")

source = source.split('^')


print(source)

result = []




for s in source:
    for c in string.punctuation:
            s = s.replace(c, "")
    result.append(s)

for _ in range(result.count('')):
    result.remove('')


        

```
