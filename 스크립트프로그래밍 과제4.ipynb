### (1) 설명
#### - w모드 open으로 알맞은 텍스트 데이터를 저장한 s.txt 파일을 생성하였다

```python
# (2)

with open("s.txt", "r") as f:
    with open("s1.txt", "w") as f1:
        line = sorted(f.readlines())
        for l in line:
            f1.write(l)

with open("s1.txt", "r") as f:
    print(f.read())
```

### (2) 설명
#### - s.txt의 첫번째 단어를 기준으로 라인별로 정렬하였다. readlines로 텍스트 파일의 라인별로 리스트를 만들어주고 그 리스트를 sorted해주면 자연스럽게 첫문자를 기준으로 정렬하기 때문에 원하는 결과를 얻을 수 있었다. 이렇게 정렬한 리스트를 s1.txt 파일에 저장하였다.

```python
# (3)

with open("s.txt", "r") as f:
    with open("s2.txt", "w") as f1:
        line = sorted(f.readlines(), key=lambda x : x[4])
        for l in line:
            f1.write(l)

with open("s2.txt", "r") as f:
    print(f.read())
```

### (3) 설명
#### - 이번에는 두번째 단어를 기준으로 정렬해야하는데, 앞 단어가 모두 3글자이기 때문에 4번째 인덱스를 기준으로 리스트를 정렬하게되면, 원하는 결과를 얻을 수 있었다. 

```python
# (4)

with open("s.txt", "r") as f:
    with open("s3.txt", "w") as f1:
        line = f.read().split()
        while line:
            tmp = " ".join(line[:3])
            line = line[3:]
            f1.write(tmp + '\n')
            
with open("s3.txt", "r") as f:
    print(f.read())
```

3. range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

  - 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
  > range(stop)<br/>
  > range(start, stop[, step])<br/>
  - 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다. 
  - 3) frange 사용 예
    - 3-1) frange(0.5)
      - [0.0, 0.1, 0.2, 0.3, 0.4]
    - 3-2) frange(1.0, 2.0)
      - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    - 3-3) frange(2.2, 4.0, 0.5)
      - [2.2, 2.7, 3.2, 3.7]

```python
def frange(*step):
    if len(step) == 1:    return [i/10 for i in range(int(step[0]*10))]
    elif len(step) == 2:  return [i/10 for i in range(int(step[0]*10), int(step[1]*10))]
    elif len(step) == 3:  return [i/10 for i in range(int(step[0]*10), int(step[1]*10), int(step[2]*10))]
    else return []
    
print(frange(0.5))
print(frange(1.0, 2.0))
print(frange(2.2, 4.0, 0.5))
```

### 코드 설명
#### 가변인자로 인자를 받고 가변인자의 길이에 따라 기능을 먼저 분류하였다. range 기능과 동일하게 동작하므로, range를 그대로 사용하는데 이때, 범위에 10을 곱해주고 리스트에 넣어줄 때 다시 10을 나누어주면 원하는 결과를 얻을 수 있다

4. 가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

  - sum() 사용 예

> sum()<br/>
>   0

> sum(1, 2)<br/>
>   3

> sum(1, 2, 3, 4, 5)<br/>
>   15

> sum(1, 5, 7, 2, -10)<br/>
>   5

```python
def sum(*argu):
    result = 0
    for i in argu:
        result += i
    return result


print(sum())
print(sum(1, 2))
print(sum(1,2,3,4,5))
print(sum(1, 5, 7, 2, -10))
```

### 코드 설명

  - 2) myinitial() 함수 사용 예
    - myinitial("as soon as possible")

```python
def myinitial(str):
    return ''.join(map(lambda x: x[0].upper(), str.split()))


myinitial("as soon as possible")
```

6. 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.
  - [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다. 

```python
def myfact(n):
    return 1 if n == 1 else n * myfact(n-1) 

myfact(2500)
```

### 코드 설명
#### 팩토리얼은 1부터 n까지의 모든 자연수를 곱해주는 것이다. 따라서 n==1이 아니라면, n을 계속 return하면서 myfact(n-1)을 계속 호출하며 곱해준다. 그리고 n==1일때 1만 return하여 재귀함수를 탈출하게되면, 원하는 결과를 얻을 수 있다.
#### [옵션]을 해보았는데 약 2500이 넘어가면서 수행이 되지 않는 걸 확인하였다.

```python
import requests
import string

def delete_tag(s, start_tag, end_tag):
    start = 0
    end = 0
    while True:
        start = s.find(start_tag)
        end = s.find(end_tag) + len(end_tag)
        if start == -1:
            break
        tmp = s.split(s[start:end], maxsplit=1)
        s = tmp[0] + '^' + tmp[1]
    return s

uml = "https://www.naver.com/"
req = requests.get(uml)
source = req.text

source = delete_tag(source, "<script", "</script>")
source = delete_tag(source, "<style", "</style>")
source = delete_tag(source, "<", ">")
source = delete_tag(source, "&", ";")
source = source.replace('\n', "^")
source = source.replace('\r', "^")
source = source.replace('\t', "^")
source = source.replace(',', "^")
source = source.replace('ㆍ', "^")
source = source.replace('…', "^")
source = source.replace('·', "^")


source = source.split('^')

for c in string.punctuation:
    source = [i.replace(c, "") for i in source]

source = ' '.join(source).split()

dic = {}

for s in source:
    dic[s] = dic.get(s, 0) + 1 
    keys = sorted(dic.keys())
    
for i in keys:
    print(i + " : " + str(dic[i]))
```
